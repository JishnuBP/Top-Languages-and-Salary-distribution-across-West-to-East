```python
from collections import Counter
```

```python
df['LanguageWorkedWith'] = df['LanguageWorkedWith'].astype(str)
```

```python
language_list = sum(df['LanguageWorkedWith'].str.split(';').tolist(), [])
```

## The working of above code:
- If you look below we can see that we will be using a Counter to count all the individual language which should be provided in a single list.
- But If you look above, df['LanguageWorkedWith'] gives us language for each respondent and its "NOT in a SINGLE LIST"
- Hence to create a single list we can use sum()
- - sum() takes only one argument if we are providing it with integers
- - sum() takes two argument if we are providing it with a string and the 2nd argument is [ ] its an empty list. This will enable us to concat different string together. 
- Now, df['LanguageWorkedWith'] needs to be converted to string so that we can use functions that belongs to string like split(' ')
- After we use split(' ') we get individual list for each respondent 
- Since sum() takes only single list as its first argument. We need to put all different list together so that we will get a single list with multiple list inside and the sum() does not care if it has a multiple list inside single list
- hence we will be using tolist() to get a single list.
- Now when we use sum() we will get a single list with no sub list inside ( it uses empty list [ ] to do it)


```python
language_list
```

```python
language_counts = Counter(language_list)
```

```python
language_counts
```

```python
type(language_counts)
```

## The working of code
- we provide a single list to the Counter so that it can count number of unique strings inside that list.
- The output that we get is a Counter object.
- And Pandas can't use a Counter object to create a data frame.
- So we need to convert a Counter object to a dict so that pandas can convert it to a data frame.
- hence we provide items() to convert a counter object to dict as shown below.
- Since counter does not comes with column names, we provide them with one as shown below.

```python
Pop_lang_df = pd.DataFrame(language_counts.items(), columns=['Language', 'Usage']) # Pop_lang = Popular language
```

```python
Pop_lang_df.sort_values('Usage', ascending = False, inplace = True)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(Pop_lang_df['Language'], Pop_lang_df['Usage'], color='skyblue')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.xlabel('Usage')
plt.ylabel('Languages')
plt.title('Language Counts')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
sal_res_df = df[['Country','ConvertedComp','Respondent']].groupby('Country').agg({'ConvertedComp':'median','Respondent':'count'})
```

```python
sal_res_df.reset_index(inplace = True)
```

```python
sal_res_df
```

```python
sal_res_df = sal_res_df.sort_values('ConvertedComp', ascending = False)
```

```python
sal_res_df
```

```python
Country = sal_res_df['Country']
Salary = sal_res_df['ConvertedComp']
Respondent = sal_res_df['Respondent']

fig, ax1 = plt.subplots()
ax1.bar(Country,Salary , color='blue', alpha=0.5, label='Median Salary')
plt.xticks(rotation=90)
ax1.tick_params(axis='y')
ax1.set_ylabel('Median Salaries in (USD)')
ax1.set_xlabel('Country')



ax2 = ax1.twinx()
ax2.plot(Country, Respondent, color='r', marker='o', linestyle='--', label='Number of Respondents')

ax2.set_ylabel('Number of Respondents')
ax2.tick_params(axis='y')

ax1.set_title('Median Salary in (USD) & Number of Respondent by Country')

ax1.legend(loc='upper right', bbox_to_anchor=(1, 1));  # For the bar plot & ';' is to hide the output '<matplotlib.legend.Legend at 0x28..m3...>' from being displayed
ax2.legend(loc='upper right', bbox_to_anchor=(1, 0.9)) ; # For the line plot

# fig.savefig('fig.png') To save the graph.
```
